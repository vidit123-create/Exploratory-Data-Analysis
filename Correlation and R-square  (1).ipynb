```python
#Let's check about the multicolinearity?

#Ans:-

       #As we know that multicolinearity means the relationship or dependency between two independent variables.
    
       #Relationship between two independent variables means that when we take the linear equation i.e y=b0+b1x then this equation is just transformed into x2=b0+b1x.
        
       #Due to which when we checked the linearity i.e.relationship or dependency between independent and dependent then we will not able to find out the exact dependency because we will see that due to relationship between two independent variables.
    
       #When we will put different values of one particular independent variable then we will see that the value of dependent variable will change as well the other independent variable will also change and due to which we will not able to see whether the dependent variable depends the most on which independent variable.
        
       #So to remove the multicolinearity we have generally used either chi-square method for categorical variables and pearson's correlaytion coefficient or scatter plot for numerical variables. 
```

```python
#What is chi-sqaure method of independence?

#Ans:-

      #Chi-square method of independence is generally used to check the association or dependency between two categorical variables using contingency matrix.
    
      #Here chi-square method generally used to check association using contingency matrix which means that we generally used contingency matrix which is generally used to classify the two variables on the basis of stasitics i.e.it is used to show how the data in two variables are classified within each other.
        
      #So generally chi-square is generally used to prepare an array which contains observed values by the use of formula:-
    
            #Denoted by x^2=Sum of all[(O-E)/E]
        
      # Now here O represents observed values,E represents expected values.
    
      #By the use of this formula chi-square is generally used to give an array which contain observed values as well as it also gives us p-value and degree of freedom.
        
      #Now how we could check about the multicolinearity between two variables using chi-square method.
    
      #Actually chi-square method generally uses the concept of p-value to check multicolinearity i.e.p-value is generally used to tell us about the deviation which occurs between two variables i.e.how much difference is there in observed value and expected value.
        
      #On the basis of that we generally get some p-value and on the basis of that p-value we check about two things:-
    
           #Null hypothesis:-Null hypothesis is a term which tell us that if there is two variables then we can't use the values of one variable to predict the value of other variables.
        
           #But as we are getting observed values by chi-square method with a p-value<0.005 then we could say that there is dependency between two variables and we are rejecting the concept of null hypothesis.
            
      #We also calculate the dregree of freedom by using formula:-
    
           #(No of rows-1)*(No of columns-1) 
```

```python
#Now we will check for the approximation betweeen actual dependent variable and predicated dependent variable by the use of r-square or coefficient of determination.

#Now r-squared is generally used to check whether how much close regressor fitted line is to the scatter plots.

#r-squared is a term which tells about the approximations between y_test and y_pred because we know that we generally train the machine i.e.make the machine learn about the data or features by passing training data.

#Once the machine have learn about the data now we give test data as x_test so if machine have learn about the data well then machine will predict some values which will be stored in a variable as y_pred.

#Then we will see how much close is y_pred with y_test because if closeness is there then accuracy will be there because residuals or outliers will be less.

#This closeness is calculated by using r-squared or coefficient of determination by using formula:-

      #Formula:-
    
              #R-sqaured=1-(RSS/TSS)
        #where RSS =Residual sum of square and TSS =Total sum of square.
        
#So if approximation is more than 50% then we could say that error is less than prediction is good means accuracy is there in the model.

#Residual are the errors which is calculated as difference between actual dependent variable and predicated dependent variable.
```
