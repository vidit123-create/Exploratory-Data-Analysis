```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('C:\\Users\\arpit\\Downloads\\Fuel_Data_(Rev.01-25-Feb-20).xlsx')
df
```

```python
#Data Cleaning
```

```python
def a(data):
    data=str(data)
    data_x=data.replace('-','')
    data_y=data_x.replace(' ','')
    data_z=data_y.lower()
    return data_z[0:4]
df['fuel_clean_vehicle']=df['VEHICLE'].apply(a)
df
```

```python
#Subsetting
```

```python
df1=df[df['fuel_clean_vehicle']!='pool']
df1
```

```python
#Read master dataset 
```

```python
df2=pd.read_excel('C:\\Users\\arpit\\Downloads\\Master.xlsx')
df2
```

```python
#Data cleaning
```

```python
def a(data):
    data=str(data)
    data_x=data.replace(' ','')
    data_y=data.replace('-','')
    data_z=data_y.lower()
    return data_z[0:4]
df2['master_clean_vehicle']=df2['VEHICLE NO'].apply(a)
df2
```

```python
#Merge the two dataframes
```

```python
df3=pd.merge(df1,df2,on='PTROCRD',how='inner')
df3
```

```python
#Now compare and find the total number of mismatch values
```

```python
df3['matched_unmatched']=df3['fuel_clean_vehicle']!=df3['master_clean_vehicle']
df3
```

```python
#Do subsetting
```

```python
x=df3[df3['matched_unmatched']==True]
x
```

```python
df4=x.groupby(['PTROCRD','fuel_clean_vehicle','master_clean_vehicle','matched_unmatched']).agg({'matched_unmatched':'count','EXPAMT':'sum'})
df4
```

```python
#Now rename the matched_unmatched
```

```python
df4.rename({'matched_unmatched':'mismatched'},axis=1,inplace=True)
df4
```

```python
#Now create a new column by assigning
```

```python
df5=df4
df5
```

```python
#Data Manipulation
```

```python
#Now sort the data_values according to Amount column
```

```python
df5.sort_values(by='EXPAMT',ascending=False)
```
