```python
# Why we try to find out correlation between two variables or many variables?

#Ans:-

      # 1.Actually correlation is generally used to find the relationship between two variables.
    
      # 2.These relationships between two variables or many variables are generally used to show how one variable changes w.r.t another variable.
        
      # 3. Actually correlation is a term which says that if two variables are correlated to each other then if there is some changes in one variable then that change will effect the other variable to which it is correlated to and in that variable also there will be some change.
    
      # 4. Actually we find correlation to see whether which is that independent variable to which the dependent variable depends the most and then we try to implement model building in that or predictions. 

# How to identify correlation?

#Ans:-

      # 1.Actually correlation are identified using graphs and several correlation methods like pearsons correlation.    
```

#### Now let's see the correlation matrix or corrlaeogram using heatmap using seaborn library:-

```python
# Why we used pearson's correlation coefficient for creating a correlation matrix?

#Ans:-

       # 1.Actually we know that as pearson's correlation coefficient is generally used to check whether there is a linear relationship between two variables or not using pearson's formula.
    
       # 2.If suppose through pearson's correlation coefficient we get a high correlation between two variables which is more than 50% than we could say that it obeys the linear equation i.e.y=b0+b1x which means that if we put values in x then we will get the same value or approx value for y.
        
       # 3.If we get a linear relationship then we could represent it with scatter plot also to see a straight line exact or with some deviation.
    
# Why we used heat map for seeing correlation matrix or corrlaeogram?

#Ans:-

       # 1.Actually Heat map is generally used to see the correlation matrix in a more proper way.
    
       # 2.Through heat map one could easily identify about the independent and dependent variable by seeing their pearson's correlation coefficient.
        
       # 3.Hence we could easily take out the independent variable and dependent variable and we could implement Linear regression model on the variables by splitting.
    
# Why we used colours in heat map?

#Ans:- 

       # 1.Since heat map will give us a better lookout for identifying correlation between variables.
    
       # 2.By using colors in the heat map we could classify which variables have negative correlation and which variables have moderate and high correlation.
        
       # 3.Here we could see that diagonal line in the heat map shows that we are checking correlation of one variable with itself and which will always be 100%. 
```

```python
fig,ax=plt.subplots(figsize=(10,10))
sns.heatmap(df1.corr(method='pearson'),annot=True,ax=ax)
```

```python
# After seeing the heat map we could easily identify that AU_MOTOR i.e.Fuel_capacity is the column which shows the high correlation of 96% or 0.96 in fraction and which means that EXPAMT i.e.Amount column is highly dependent on this AU_MOTOR variable.
```

```python
#Since we know that heatmap is used to identify the independent variable to which dependent variable depends the most i.e a type of correlation we could find from the heatmap between variables by just seeing the statistics derived from the heat map.

# Here we could see that we have AU_MOTOR i.e fuel_capacity is the variable to which the EXPAMT i.e Amount column depends the most because we could see that we have correlation of more than 0.5 or 50% so we will take this both variable as independent and dependent variable.
```

```python
#Independent variable
```

```python
x=df[['AU_MOTOR']]
x
```

```python
#Dependent variable
```

```python
y=df[['EXPAMT']]
y
```

```python
#Now split the data into training and test data.

#Here we will split the dependent variable also because we know that we are using supervised learning.

#We know that supervised learning means where resultant variable is present in the dataset and so while doing splitting we make the machine learn about the variables.

#So we have to make machine learn about the resultant variable then only machine will know what to predict.
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=0)
```

```python
#So here we have given approx 10% i.e 0.10 of the data into test data and 90% i.e.0.90 of the data into training data.

#This means that out of 1954 total rows 1758 rows are contained by training data and 196 rows are contained by the test data.
```

```python
#Now we will apply the linear regression model.
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#Now we will see the predicated dependent variable values by passing independent test data i.e.x_test.
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
#Now we will calculate the r2 or R-square metrics i.e coefficient of determination or least square method to see approximation between the observed dependent variable i.e.y_test and predicated dependent variable i.e.y_pred.
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
#Since we know that r2 or coefficient of determination is generally used to see whether how much is the regressor fitted line is close to the scatter plot.

#This means that how much is the observed dependent variable is close to the predicated dependent variable and this approximation is generally calculated by r2 or least square method.

#In this we have high approximation or we could say that we have more closeness in both observed dependent variable and predicated variable with 0.93 or 93% approximation and 7% error rate.
```

```python
#Now let's calculate the residual i.e. the error between observed dependent variable and predicated dependent variable.

#If we will calculate the mean square error then we will get approx 7% error rate.
```

```python
residual=y_test-y_pred
residual
```

```python
#Now let's see the regressor fitted line is how much close to the scatter plot by graph.
```

```python
#Graph:-
```

```python
fig,ax=plt.subplots(figsize=(8,6))
ax.scatter(x_train,y_train,color='r',label='TRAINING_DATA')
ax.plot(x_test,y_pred,color='b',label='TEST_DATA')
plt.title('AU_MOTOR VS EXPAMT')
plt.legend()
plt.xlabel('AU_MOTOR')
plt.ylabel('EXPAMT')
plt.show()
```
